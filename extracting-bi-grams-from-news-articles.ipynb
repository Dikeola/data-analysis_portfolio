```python
df= pd.read_csv('/kaggle/input/news-articles/Articles.csv', encoding='hp_roman8')
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(max_features=100, stop_words='english')

csr = count.fit_transform(df.Article)
```

```python
df_2 = pd.DataFrame(csr.toarray(), columns=count.get_feature_names_out()).add_prefix('Count_')
```

```python
df_2.head()
```

```python
count_2 = CountVectorizer(ngram_range=(2,2))
```

```python
csr_2 = count_2.fit_transform(df.Article)
```

```python
pd.DataFrame(csr_2.toarray(), columns=count_2.get_feature_names_out())
```

```python
cv_trigram = CountVectorizer(max_features=100, stop_words='english', ngram_range=(3,3))
csr_trigram = cv_trigram.fit_transform(df.Article)
```

```python
df_tri = pd.DataFrame(csr_trigram.toarray(), columns=cv_trigram.get_feature_names_out()).add_prefix('Count_')
```

```python
df_tri.sum().sort_values(ascending=False)
```

```python
cv_bi = CountVectorizer(max_features=100, stop_words='english', ngram_range=(2,2))
csr_bi = cv_bi.fit_transform(df.Article)
```

```python
df_bi = pd.DataFrame(csr_bi.toarray(), columns=cv_bi.get_feature_names_out())
```

```python
df_bi.sum().sort_values(ascending=False)
```
