```python
df = pd.read_csv('/kaggle/input/celebrity-deaths/celebrity_deaths_4.csv', encoding='mac_latin2')
```

```python
df.info()
```

```python
df.head()
```

```python
df_clean = df.dropna()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(max_features=100, ngram_range=(2,2), stop_words='english')
```

```python
csr_bi = count.fit_transform(df_clean.famous_for)
```

```python
df_bi = pd.DataFrame(csr_bi.toarray(),columns=count.get_feature_names_out())
```

```python
df_bi.sum().sort_values(ascending=False)
```

```python
count_tri = CountVectorizer(max_features=100, ngram_range=(3,3), stop_words='english')
csr_tri = count_tri.fit_transform(df_clean.famous_for)

df_tri = pd.DataFrame(csr_tri.toarray(), columns=count_tri.get_feature_names_out())
```

```python
df_tri.sum().sort_values(ascending=False)
```

```python
count_quad = CountVectorizer(max_features=100, ngram_range=(4,4), stop_words='english')
csr_quad = count_quad.fit_transform(df_clean.famous_for)
df_quad = pd.DataFrame(csr_quad.toarray(), columns=count_quad.get_feature_names_out())
```

```python
df_quad.sum().sort_values(ascending=False)
```
