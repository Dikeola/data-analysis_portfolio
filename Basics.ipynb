There are 5,699 rows and 41 columns.
All columns are of data-type object except for *Iceland Krona, Romanian leu and Turkish lira; which are float64.

DATA CLEANING

```python
exchange_rates.isnull().sum()
```

There are several null values.

```python
exchange_rates.rename(columns={'[US dollar ]':'US_dollar','Period\\Unit:':'Time'},inplace=True)
```

```python
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar=exchange_rates[['Time','US_dollar']]
euro_to_dollar['US_dollar'].value_counts()
```

```python
exchange_rates[exchange_rates['US_dollar']=='-']['US_dollar'].value_counts()
```

There are 62 fields in the US_dollar column with '-' character.

Removing rows with them in the code below.

```python
exchange_rates['US_dollar']=exchange_rates[~(exchange_rates['US_dollar']=='-')]['US_dollar']
```

```python
#converting the US_dollar column to float type
exchange_rates['US_dollar']=exchange_rates['US_dollar'].astype(float)
```

```python
import datetime as dt
import matplotlib.style as style
import matplotlib.pyplot as plt
%matplotlib inline 
# Enables Jupyter to display graphs
```

creating a new column 'rolling_mean' that stores the rolling mean with a moving window of 30 days. 
One major characteristic is that it noise in the plotted graph, consequently generating a smoother graph.

```python
exchange_rates['rolling_mean']=exchange_rates['US_dollar'].rolling(30).mean()
```

plotting a line graph that highlights the period of the covid crisis(2020-)

```python
style.use('ggplot')
fig,ax=plt.subplots(figsize=(4,7))
ax.plot(exchange_rates['Time'],exchange_rates['rolling_mean'],color='#af0b1e',alpha=0.05)
ax.grid(visible=False)
#ax.set_xticklabels([])
#ax.set_yticklabels([])
for spine in ['left','right','top','bottom']:
    ax.spines[spine].set_visible(False)
ax.tick_params(left=False,bottom=False)
ax.plot(exchange_rates[exchange_rates['Time'].dt.year>=2020]["Time"],exchange_rates[exchange_rates['Time'].dt.year>=2020]['rolling_mean'],color='#af0b1e',linewidth=2)
plt.xlabel('Year',weight='bold',size=14)
plt.title('US Dollar Exchange Rate')
plt.show()
```
