```python
df = pd.read_csv('/kaggle/input/covid19-vaccine-articles/COVID-19_Vaccine.csv')
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(min_df=.2, max_df=.8)
csr = count.fit_transform(df.ArticleText)
```

```python
df_2 = pd.DataFrame(csr.toarray(),columns=count.get_feature_names_out()).add_prefix('Count_')
comb_df = pd.concat([df,df_2], axis=1, sort=False)
```

```python
comb_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer()

csr_mat = tf.fit_transform(df.ArticleText)
```

```python
df_tf = pd.DataFrame(csr_mat.toarray(), index=df['Author(s)'])
```

```python
df_tf.head()
```

```python
target = df_tf.loc['Brenda Goodman']
sim = df_tf.dot(target)

sim.nlargest()
```
