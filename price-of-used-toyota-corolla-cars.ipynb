There is a difference between the average price of automatic cars and manual cars. Further analysis would be done to check whether or not it is by chance.

```python
from scipy.stats import f_oneway,ttest_ind
```

```python
tstat,pval = ttest_ind(df.loc[df.automatic==0,'price'],df.loc[df.automatic==1,'price'])
#Null Hypothesis: There is no association between automatic cars and price
#Alternative Hypothesis: There is an association between automatic cars and price; Whether or not the car is automatic influences the price.
if pval <0.05:
    print('Reject Null Hypothesis. There is an association between price and automatic cars')
else:
    print('Accept Null Hypothesis. There is no association between price and automatic cars. Difference is by chance.')
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
features = []
```

Going to add features that influence the price of cars; whether or not the car is automatic has no influence so it wont be added.

```python
df.fuel_type.value_counts()
```

```python
fstat,pval = f_oneway(df.loc[df.fuel_type=='Petrol','price'],df.loc[df.fuel_type=='Diesel','price'],df.loc[df.fuel_type=='CNG','price'])
#Null Hypothesis: There is no association between any of the fuel types and price.
#Alternative Hypothesis: There is an association between atleast two fuel types and price.
if pval <0.05:
    print('Reject Null Hypothesis. There is atleast an association between two fuel types.')
else:
    print('Accept Null Hypothesis. There is no association between any of the fuel types.')
```

```python
result = pairwise_tukeyhsd(df.price,df.fuel_type)
print(result)
```

There is an between all the fuel types; all three possible combinations.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
df[features].info()
```

```python
df = pd.get_dummies(data=df,columns=['fuel_type'])
```

```python
df.head()
```

```python
features.remove('fuel_type')
X = df[features+['fuel_type_CNG','fuel_type_Diesel','fuel_type_Petrol']]
y = df.price

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.25,random_state=40)
model1 = LinearRegression()
model1.fit(X_train,y_train)
```

```python
ypred = model1.predict(X_test)
r2_score(y_test,ypred)
```

The model can account for 73% of the variation in prices; it is 73% accurate.
