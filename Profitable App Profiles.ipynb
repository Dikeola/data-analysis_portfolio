```python
reviews_max = {}
def conv(x):#function to convert reviews that have letters M and K
    if 'M' in x:
        x= x.strip('M')
        return float(x)*1000000
    elif 'K' in x:
        return float(x)*1000
    return float(x)   
for row in google[1:]:
    name = row[0]#points to the app name
    n_reviews = conv(row[3])#points to the app reviews
    if name in reviews_max and reviews_max[name]<n_reviews:
        reviews_max[name]=n_reviews
    elif name not in reviews_max:
        reviews_max[name]=n_reviews    
```

```python
print(len(reviews_max))
```

Created two new lists;
"android_clean" and "already_added"
one to store the new cleaned dataset
and the other to store the names of apps already added to the new dataset.

android_clean list is used to store lists.

```python
android_clean = []# to store cleaned dataset
already_added = []# to store app names
for row in google[1:]:
    name = row[0]# name of the app
    n_reviews = conv(row[3])#reviews column
    if n_reviews == reviews_max[name] and name not in already_added:
       android_clean.append(row)
       already_added.append(name)        
```

To detect foreign characters, create a function that checks if any of the characters is greater than 127; commonly used characters in English are within 0-127 range

```python
def check_foreign(strr):#function checks if there is a foreign character in input string
  for i in strr:
    if ord(i)>127:
        return False
  return True  
```

Removing Non-English Apps: PART 2

we saw that the function couldn't correctly identify certain English app names like 

'Docs To Go‚Ñ¢ Free Office Suite' and 'Instachat üòú'

This is because emojis and characters like ‚Ñ¢ fall outside the ASCII range and have corresponding numbers over 127.

```python
print(ord('üòú'))
print(ord('‚Ñ¢'))
```

If we're going to use the function we've created, we'll lose useful data since many English apps will be incorrectly labeled as non-English.

To minimize the impact of data loss, we'll only remove an app if its name has more than three characters with corresponding numbers falling outside the ASCII range.

```python
def check_foreign(strr):#correcting the function to check if more than 3 characters fall outside the Ascii range
    count = 0#variable to count number of outliers; chars that fall outside the Ascii range
    for i in strr:
        if ord(i) > 127:
           count+=1
        if count==3:
            return False
    return True   
```

```python
print(check_foreign('Docs To Go‚Ñ¢ Free Office Suite'))
print(check_foreign('Instachat üòú'))
print(check_foreign('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
```

We'll build two functions we can use to analyze the frequency tables:

--One function to generate frequency tables that show percentages

--Another function we can use to display the percentages in a descending order

```python
def freq_table(dataset,index,header=False):
    freq_table={}
    if header:
        dataset=dataset[1:]
    for row in dataset:
        cat = row[index]# category of focus
        if cat in freq_table:
            freq_table[cat]+=1
        elif cat not in freq_table:
            freq_table[cat]=1
    for cats in freq_table:#converting frequencies to percentages   
       freq_table[cats]=(freq_table[cats]/len(dataset))*100 
    return freq_table 
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

```python
display_table(free_apple_apps,11)
print('\n')
display_table(free_google_apps,9)
display_table(free_google_apps,1)
```

**Most Popular Apps by Genre on the App Store.

*One way to find out what genres are the most popular (have the most users) is to calculate the average number of installs for each app genre.

 *For the Google Play data set, we can find this information in the **Installs** column, but this information is missing for the App Store data set.
 As a workaround, we'll take the total number of user ratings as a proxy, which we can find in the **rating_count_tot** app.

**Let's start with calculating the average number of user ratings per app genre on the App Store. To do that, we'll need to do the following:

-- Isolate the apps of each genre
-- Add up the user ratings for the apps of that genre
-- Divide the sum by the number of apps belonging to that genre (not by the total number of apps)

Average ratings of genre in the App Store

```python

genre_ratings = {}  # Dictionary to store genre ratings

for row in free_apple_apps:
    genre_app = row[11]  # Save the app genre
    ratings = float(row[5])  # Save the number of ratings of the app
    
    if genre_app in genre_ratings:
        genre_ratings[genre_app]['total'] += ratings
        genre_ratings[genre_app]['count'] += 1
    else:
        genre_ratings[genre_app] = {'total': ratings, 'count': 1}

for genre, data in genre_ratings.items():
    avg = data['total'] / data['count']
    print("{} average rating is {:.2f}".format(genre, avg))
    
```
