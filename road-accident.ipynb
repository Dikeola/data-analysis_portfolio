```python
df = pd.read_csv('/kaggle/input/car-accident-dataset/Road Accident Data.csv')
```

```python
df.columns = map(str.lower, df.columns)
```

```python
df.drop(columns = 'carriageway_hazards', inplace=True)
```

```python
from datetime import datetime as dt
```

```python
df.time = pd.to_datetime(df.time, format='%H:%S')
```

```python
df.rename(columns={'accident date':'accident_date'}, inplace=True)
df.accident_date = pd.to_datetime(df.accident_date)
```

```python
df.info()
```

```python
df.drop(columns=['longitude','latitude'], inplace=True)
```

```python
df.groupby('accident_severity')['number_of_casualties'].mean()
```

```python
df.vehicle_type.value_counts()
```

```python
df.pivot_table(index='day_of_week',columns='accident_severity',values='number_of_casualties', aggfunc='mean')
```

```python
df.pivot_table(index='day_of_week',columns='urban_or_rural_area',values='number_of_casualties', aggfunc='mean', margins='sum')
```

```python
df.pivot_table(index='day_of_week',columns='urban_or_rural_area',values='number_of_casualties', aggfunc=(lambda a: a.max()))
```

```python
df['vehicle_to_casualty_ratio'] = df.number_of_vehicles/df.number_of_casualties
df.pivot_table(index='day_of_week',columns='urban_or_rural_area',values='vehicle_to_casualty_ratio', aggfunc='max')
```

```python
df.info()
```

```python
df['hour_of_day'] = pd.to_numeric(df.time.dt.strftime('%H'), errors='coerce')
```

```python
df['hour_of_day'].max()
```

```python
df['hour_of_day'].min()
```

```python
bins = [0,9,12,15,18,21,23]
labels = ['0-9','9-12','12-15','15-18','18-21','21-23']
df['time_category'] = pd.cut(df.hour_of_day,bins=bins,labels=labels)
```

```python
df['time_category'].value_counts(normalize=True)*100
```

```python
df.road_type.nunique()
```

```python
pd.crosstab(index=[df.day_of_week,df.time_category],columns =df.road_type,values=df.number_of_casualties, aggfunc='max',colnames=['Road Type'])
```

```python
pd.crosstab(index=[df.day_of_week,df.time_category],columns =df.road_type,values=df.number_of_casualties, aggfunc='mean',colnames=['Road Type'])
```

```python
df.groupby('vehicle_type')[['number_of_casualties','vehicle_type']].agg({'number_of_casualties':'sum','vehicle_type':'size'}).rename(columns={'vehicle_type':'Count'})
```

```python
vehicles = df.groupby('vehicle_type')[['number_of_casualties','vehicle_type']].agg({'number_of_casualties':'sum','vehicle_type':'size'}).rename(columns={'vehicle_type':'Count'})
```

```python
vehicles
```

```python
vehicles.apply(lambda a: a.number_of_casualties/a.Count,axis=1)
```

Pedal cycle and Ridden horse would be dropped because there isnt much record on both.

```python
vehicles = vehicles.reset_index()
```

```python
vehicles
```

```python
vehicles.drop(index=vehicles[vehicles.vehicle_type.isin(['Pedal cycle','Ridden horse'])].index, inplace=True)
```

```python
vehicles.set_index('vehicle_type').apply(lambda a: a.number_of_casualties/a.Count,axis=1).sort_values(ascending=False).iloc[:5]
```

Agricultural Vehicles have the highest number of casualties per vehicle.
