```python


copy=autos.columns
print(copy)#display initial column names
new_cols=[]
for col in copy:# iterating through the column_names
    if col=='yearOfRegistration':
        new_cols.append('registration_year')
    elif col=='monthOfRegistration':
        new_cols.append('registration_month')
    elif col=='notRepairedDamage':
        new_cols.append('unrepaired_damage')
    elif col=='dateCreated':
        new_cols.append('ad_created')
    else:
        new_cols.append(col.lower())
                       
autos.columns=new_cols
autos.head()                       
                       
                       
```

Exploring Price by Brand

i have chosen to choose the top 20 brands.

```python
top_20= autos['brand'].value_counts(normalize=True).head(20)
```

```python
top_20.index[1]
```

```python
brand_price_agg={}
brand_mileage={}
for brand in top_20.index:
    price_mean = autos.loc[autos['brand']==brand,'price'].mean()
    mileage_mean=autos.loc[autos['brand']==brand,'odometer_km'].mean()
    brand_price_agg[brand]=round(price_mean,2)
    brand_mileage[brand]=round(mileage_mean,2)
```

```python
print(brand_price_agg)
```

```python
print('{} has the highest price average of {}'.format(max(brand_price_agg),brand_price_agg[max(brand_price_agg)]))
```

```python
print('{} has the lowest price average of {}'.format(min(brand_price_agg),brand_price_agg[min(brand_price_agg)]))
```

```python
print(brand_mileage)
```

```python
mile = pd.Series(brand_mileage)#Convert milege to series
```

```python
print(mile)
```

```python
price_avg=pd.Series(brand_price_agg)
```
