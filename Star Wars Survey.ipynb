Convert the Have you seen any of the 6 films in the Star Wars franchise? column to the Boolean type.
Convert the Do you consider yourself to be a fan of the Star Wars film franchise? column to the Boolean type.

```python
yes_no = {'Yes':True,'No':False}
s='Have you seen any of the 6 films in the Star Wars franchise?'
star_wars[s]=star_wars[s].map(yes_no)
s2='Do you consider yourself to be a fan of the Star Wars film franchise?'
star_wars[s2] = star_wars[s2].map(yes_no)
```

```python
star_wars[s2].value_counts()
```

```python
star_wars[s].value_counts()
```

```python
yes_no2 = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.NaN: False,
    'Star Wars: Episode II Attack of the Clones':True,
    'Star Wars: Episode III Revenge of the Sith':True,
    'Star Wars: Episode IV A New Hope':True,
    'Star Wars: Episode V The Empire Strikes Back':True,
    'Star Wars: Episode VI Return of the Jedi':True
}
star_wars = star_wars.rename(columns={'Unnamed: 4':'seen_1', 'Unnamed: 5':'seen_2', 'Unnamed: 6':'seen_3', 'Unnamed: 7':'seen_4',
       'Unnamed: 8':'seen_5'
})
cols = star_wars.columns[3:9]
for col in cols:
    star_wars[col]=star_wars[col].map(yes_no2)
```

```python
star_wars['seen_5'].value_counts()
```

Comverting movies seen columns to float type to allow for further calculations/analysis.

```python
star_wars[star_wars.columns[9:15]]=star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns={'Unnamed: 10':'ranking_1','Unnamed: 11':'ranking_2','Unnamed: 12':'ranking_3','Unnamed: 13':'ranking_4','Unnamed: 14':'ranking_5'})
```
